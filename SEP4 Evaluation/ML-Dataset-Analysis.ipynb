# Machine Learning: Dataset analysis, SEP4.
This notebook performs a simple data analysis of the collected sensor data samples at the time of 20/05 15:00.

Responses were collected from the https://sep4api.azure-api.net/api/iot/sample endpoint.

Disclaimer: LLM's were used in the generation of this code.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load JSON data
with open('SEP4 Evaluation/response-sep4api-azure-api-net-api-iot-s.json', 'r') as f:
    raw_data = json.load(f)

# Convert JSON list of SampleDTOs into a DataFrame
samples = [item['SampleDTO'] for item in raw_data['list']]
df = pd.DataFrame(samples)

# Display the first few rows to confirm structure
display(df.head())

# Count total number of entries
total_entries = len(df)
print(f"Total entries: {total_entries}")

# Count complete data entries (no null values)
complete_entries = df.dropna().shape[0]
print(f"Complete entries (no nulls): {complete_entries}")

# Plot bar charts for count distributions of key variables
key_fields = ['soil_humidity', 'air_humidity', 'light_value', 'air_temperature']

for key in key_fields:
    # Drop NaNs and convert to int
    valid_values = df[key].dropna().astype(int)

    # Build full value range between min and max of non-null entries
    if not valid_values.empty:
        full_range = range(valid_values.min(), valid_values.max() + 1)
        value_counts = valid_values.value_counts().reindex(full_range, fill_value=0).sort_index()

        plt.figure(figsize=(16, 4))
        plt.bar(value_counts.index, value_counts.values, color='steelblue')

        if key == 'light_value':
            plt.ylim(top=25)

        plt.title(f"Bar Chart: Distribution of '{key}' Values")
        plt.xlabel(f"{key} Value")
        plt.ylabel("Number of Samples")
        plt.grid(axis='y')
        plt.tight_layout()
        plt.show()
    else:
        print(f"No valid data for '{key}' to plot.")
```
