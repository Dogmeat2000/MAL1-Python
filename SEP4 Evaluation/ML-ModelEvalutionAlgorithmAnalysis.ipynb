# Machine Learning: Model Evaluation Algorithm Analysis
In this Notebook, the model evaluation algorithm in use by the C#-Based Prediction-Build-Service, is analysed and evaluated.

The analysis consists of:
1. Generating a set of model_scores, one for each possible variation of the (rmse, r2, cross_val_splits) combination.
2. Using this generated data set to display a 3D scatter plot, as a means of visualizing how variations in the variables affect the model scoring.
3. Identifying the optimal variables for the best possible model score and confirming if this aligns with the intentions behind the evaluation.

Since our Machine Learning backend only utilizes Linear Regression and Random Forest models, the only performance metrics made available in each models metadata is:

- RMSE (Root Squared Mean Errors): This is a representation of how far from a precise evaluation (in minutes) the model is. I.e. a RMSE of 100 means that the model on average will make a prediction that lies about 100 minutes from the true target.
- R2: Is a metric that is used to explain how much of the variance the model accounts for. This is especially useful for identifying underfitted (below 60%) or overfitted models (above 90%).
- Cross_val_splits: Indicates how thorough the cross validation has been, while training the model. Low values can indicate that the model hasn't been thoroughly validated, which can cause inflated R2 values.
- Alpha: Indicates the regularization strength applied (in either Ridge or Lasso). Larger values penalize large coefficients more heavily, reducing overfitting - but can lead to underfitting. This metadata metric is NOT evaluated as part of the model scoring, as it is better suited for model tuning - rather than comparison between models.

Disclaimer: LLM's were used to support in the performance of this analysis (mainly with the code)

```python
import numpy as np
import pandas as pd
from itertools import product
import matplotlib.pyplot as plt
```

```python
# The ComputeModelScore method converted from C# to Python:
# It rewards better variable values and applies a penalty towards lower values.
def compute_model_score(rmse, r2, cross_val_splits):
    max_realistic = 10080
    rmse_limit   = np.sqrt(max_realistic)
    r2_lo, r2_hi = 0.7, 0.9
    w_rmse, w_r2 = 0.6, 0.4
    low_cv_pen   = 0.25

    # RMSE contribution:
    if rmse > rmse_limit:
        score_rmse = -((rmse / max_realistic) * 100)
    else:
        score_rmse = 100 - ((rmse / max_realistic) * 100)

    # R² contribution:
    if r2 > r2_hi:
        score_r2 = -((r2 - r2_hi) / (1 - r2_hi)) * 100 * 2
    elif r2 >= r2_lo:
        score_r2 = ((r2 - r2_lo) / (r2_hi - r2_lo)) * 100
    else:
        score_r2 = -((r2_lo - r2) / r2_lo) * 100

    # combined:
    score = score_rmse * w_rmse + score_r2 * w_r2

    # CV penalty:
    if cross_val_splits < 5:
        score = score * (1 + low_cv_pen) if score < 0 else score * low_cv_pen

    return score
```

## 1. Generating a dataset containing possible variations of rmse, r2 and cross_val_splits within a reasonable range.
A dataset is generated within these ranges:
- RMSE            : From 0 to 750 (in increments of 5)
- R2              : From 0 to 1 (in increments of 0.025)
- Cross_val_splits: From 0 to 8 (in increments of 1).

```python
rmse_range = range(0, 750, 5)
r2_range = np.round(np.arange(0.0, 1.01, 0.025), 2)
cv_range = range(0, 8, 1)

combinations = product(rmse_range, r2_range, cv_range)

data = []
for rmse, r2, cv in combinations:
    score = compute_model_score(rmse, r2, cv)
    data.append([rmse, r2, cv, score])

df = pd.DataFrame(data, columns=["rmse", "r2", "cross_val_splits", "model_score"])

# Display the last few variables, to see how they are formatted (and how many data entries there are):
display(df.tail());
```

## 2. Visualizing the dataset in 3D:
The data set is visualized in 3D, with a comparison to model_score, so that these intentions can be confirmed:
1. That higher R2, up until 0.9 (90%), generally corresponds with improved model_scores.
2. That lower RMSE generally correpsonds with improved model_scores
3. That Cross_Val_Splits 5 and above generally correspond with improved model_scores.


```python
# Downsample for plotting:
plot_df = df.sample(n=3000, random_state=22)

X = plot_df["rmse"].values
Y = plot_df["r2"].values
Z = plot_df["cross_val_splits"].values
C = plot_df["model_score"].values

# 3D scatter colored by model_score:
fig = plt.figure(figsize=(22, 16))
ax  = fig.add_subplot(111, projection='3d')
sc  = ax.scatter(X, Y, Z, c=C, cmap='viridis', s=12, alpha=0.8)

ax.set_xlabel("RMSE")
ax.set_ylabel("R²")
ax.set_zlabel("CV splits")
ax.set_title("3D: RMSE vs R² vs CV splits → model_score")
cbar = plt.colorbar(sc, pad=0.1)
cbar.set_label("model_score")

# Highlight best point:
best_idx = plot_df["model_score"].idxmax()
best = plot_df.loc[best_idx]

ax.scatter(
    best["rmse"], best["r2"], best["cross_val_splits"],
    color='red', s=200, marker='*', label="Best score"
)
ax.legend()

plt.show()

# Print best combination:
print(f"Best score = {best['model_score']:.2f}")
print(f" → RMSE  = {best['rmse']:.2f}")
print(f" → R²    = {best['r2']:.2f}")
print(f" → CV    = {int(best['cross_val_splits'])}")
```

### Based on the above 3d scatter plot, we see the following trends:
1. R2: As the R2 increases, from 0 towards 1.0 the model_score generally also increases. Until R2 reaches 0.9, then it drastically falls again.
2. RMSE: As the RMSE decreases towards 0, the model_score generally increases.
3. Cross_Val_Splits: Values lower than 5 generally produce worse model_score, than value at 5 or above.

These observations correspond with our intentions:
1. That higher R2, up until 0.9 (90%), generally corresponds with improved model_scores.
2. That lower RMSE generally correpsonds with improved model_scores
3. That Cross_Val_Splits 5 and above generally correspond with improved model_scores.

For even further verification, we display the 3d Scatterplot from multiple angles, in order to get an improved visualization of how each variable corresponds with the model_score.

```python
# Define 3 angles (elev, azim): front, side, top
view_angles = [
    ("Front View", 30, 60),     # default view
    ("Side View", 30, 0),       # rotate around z-axis
    ("Top View", 90, 0),        # top-down
    ("Back View", 180, 0),       # back
]

fig = plt.figure(figsize=(22, 10))

for i, (title, elev, azim) in enumerate(view_angles, 1):
    ax = fig.add_subplot(1, 4, i, projection='3d')
    sc = ax.scatter(X, Y, Z, c=C, cmap='viridis', s=5, alpha=0.6)

    ax.view_init(elev=elev, azim=azim)
    ax.set_xlabel("RMSE")
    ax.set_ylabel("R²")
    ax.set_zlabel("CV splits")
    ax.set_title(title)

# Shared colorbar
fig.subplots_adjust(right=0.9, wspace=0.3)
cbar_ax = fig.add_axes([0.93, 0.15, 0.015, 0.7])
cbar = fig.colorbar(sc, cax=cbar_ax)
cbar.set_label("model_score")

plt.suptitle("Model Score Visualization from Multiple Perspectives", fontsize=16)
plt.show()
```

These observations correspond with our intentions:
