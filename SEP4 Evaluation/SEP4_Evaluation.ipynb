```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Re‑implemented scoring function
def compute_model_score(rmse, r2, cv):
    max_realistic = 60 * 24 * 7             # 10080
    rmse_limit    = np.sqrt(max_realistic)  # ~100.4
    r2_low, r2_high = 0.7, 0.9
    w_rmse, w_r2   = 0.6, 0.4
    low_cv_penalty = 0.25

    # RMSE component
    if rmse > rmse_limit:
        score_rmse = -((rmse / max_realistic) * 100)
    else:
        score_rmse = 100 - ((rmse / max_realistic) * 100)

    # R² component
    if r2 > r2_high:
        score_r2 = -(((r2 - r2_high)/(1 - r2_high)) * 100 * 2)
    elif r2 >= r2_low:
        score_r2 = ((r2 - r2_low)/(r2_high - r2_low)) * 100
    else:
        score_r2 = -(((r2_low - r2)/r2_low) * 100)

    score = score_rmse * w_rmse + score_r2 * w_r2

    # CV penalty if splits < 5
    if cv < 5:
        score = score * (1+low_cv_penalty) if score < 0 else score * low_cv_penalty

    return score

# build a grid of points
rmse_vals  = np.linspace(0, 200, 20)
r2_vals    = np.linspace(0, 1.0, 20)
cv_splits  = np.arange(1, 6)

pts = []
for cv in cv_splits:
    for r in rmse_vals:
        for r2 in r2_vals:
            pts.append((r, r2, cv, compute_model_score(r, r2, cv)))

pts = np.array(pts)
X, Y, Z, C = pts[:,0], pts[:,1], pts[:,2], pts[:,3]

# plot
fig = plt.figure(figsize=(10,8))
ax  = fig.add_subplot(111, projection='3d')
sc  = ax.scatter(X, Y, Z, c=C)
ax.set_xlabel("RMSE")
ax.set_ylabel("R²")
ax.set_zlabel("CV splits")
ax.set_title("3D: RMSE vs R² vs CV splits → Score")
plt.colorbar(sc, ax=ax, label="Score")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401
import pandas as pd
from pandas.plotting import parallel_coordinates

# --- 1) Python version of ComputeModelScore ---

def compute_model_score(rmse, r2, cross_val_splits):
    max_realistic = 10080
    rmse_limit   = np.sqrt(max_realistic)
    r2_lo, r2_hi = 0.7, 0.9
    w_rmse, w_r2 = 0.6, 0.4
    low_cv_pen   = 0.25

    # RMSE contribution
    if rmse > rmse_limit:
        score_rmse = -((rmse / max_realistic) * 100)
    else:
        score_rmse = 100 - ((rmse / max_realistic) * 100)

    # R² contribution
    if r2 > r2_hi:
        score_r2 = -((r2 - r2_hi) / (1 - r2_hi)) * 100 * 2
    elif r2 >= r2_lo:
        score_r2 = ((r2 - r2_lo) / (r2_hi - r2_lo)) * 100
    else:
        score_r2 = -((r2_lo - r2) / r2_lo) * 100

    # combined
    score = score_rmse * w_rmse + score_r2 * w_r2

    # CV penalty
    if cross_val_splits < 5:
        score = score * (1 + low_cv_pen) if score < 0 else score * low_cv_pen

    return score

# --- 2) Build a grid of metric values and compute scores ---

# define ranges
rmse_vals = np.linspace(0, 200, 50)   # RMSE from 0 up to 200
r2_vals   = np.linspace(0.0, 1.0, 50) # R² from 0 to 1
cv_vals   = np.array([3, 5, 7])       # example CV splits

# allocate
scores = np.zeros((len(rmse_vals), len(r2_vals), len(cv_vals)))

for i, rmse in enumerate(rmse_vals):
    for j, r2 in enumerate(r2_vals):
        for k, cv in enumerate(cv_vals):
            scores[i, j, k] = compute_model_score(rmse, r2, cv)

# --- 3) 3D Surface at a fixed CV split ---

fixed_cv = 5
k = np.argmin(np.abs(cv_vals - fixed_cv))

RMSE, R2 = np.meshgrid(rmse_vals, r2_vals, indexing='ij')
Z = scores[:, :, k]

fig = plt.figure(figsize=(10,7))
ax  = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(RMSE, R2, Z, cmap='viridis', linewidth=0, antialiased=True)
ax.set_xlabel("RMSE")
ax.set_ylabel("R²")
ax.set_zlabel("Score")
ax.set_title(f"Score Surface @ CV = {fixed_cv}")
fig.colorbar(surf, shrink=0.5, aspect=10, label="Score")
plt.show()

# --- 4) Contour heatmaps for each CV split ---

fig, axes = plt.subplots(1, len(cv_vals), figsize=(5*len(cv_vals), 4), constrained_layout=True)
for ax, cv, idx in zip(axes, cv_vals, range(len(cv_vals))):
    cf = ax.contourf(rmse_vals, r2_vals, scores[:, :, idx].T, levels=20, cmap='plasma')
    ax.set_title(f"CV = {cv}")
    ax.set_xlabel("RMSE")
    ax.set_ylabel("R²")
fig.colorbar(cf, ax=axes, orientation='vertical', label='Score')
plt.show()

# --- 5) Parallel‑coordinates highlighting the best score ---

# flatten into DataFrame
df = pd.DataFrame({
    'rmse': np.repeat(rmse_vals, len(r2_vals)*len(cv_vals)),
    'r2':   np.tile(np.repeat(r2_vals, len(cv_vals)), len(rmse_vals)),
    'cv':   np.tile(cv_vals, len(rmse_vals)*len(r2_vals)),
    'score': scores.ravel()
})

# mark best
best_idx = df['score'].idxmax()
df['label'] = 'other'
df.loc[best_idx, 'label'] = 'best'

plt.figure(figsize=(10,4))
parallel_coordinates(df, 'label', cols=['rmse','r2','cv'],
                     color=['#cccccc','#d62728'], alpha=0.3)
plt.title("Parallel‑Coordinates: best model in red")
plt.ylabel("Value (scaled)")
plt.show()
```
